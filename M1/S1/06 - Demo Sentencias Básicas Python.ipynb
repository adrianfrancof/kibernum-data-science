### 6. Sentencias Condicionales

```python
edad = 55
```

```python
if edad >= 18:
    print('Eres mayor de edad, puedes pasar')
    
print('Hasta Pronto!')
```

```python
if edad >= 18:
    print('Eres mayor de edad')
else:
    print('Eres menor de edad')
```

```python
if edad >= 18 and edad < 40:
    print('Eres Joven')
elif edad >= 40 and edad < 50:
    print('Adulto Jovensaurio')
elif edad >= 50 and edad < 60:
    print('Adulto Fome')
elif edad >= 60:
    print('Adulto Mayor')
else:
    print('Eres menor de edad')
```
